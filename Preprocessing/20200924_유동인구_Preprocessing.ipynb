```python
calendar_df = pd.read_csv('만든데이터/calendar.csv')
calendar_df.head()
```

```python
calendar_dict = calendar_df.set_index('date').to_dict('index')
df['is_holiday'] = [calendar_dict[date]['is_holiday'] or calendar_dict[date]['is_weekend'] for date in df['STD_YMD']]
```

```python
df[::2004]
```

```python
flow_time_df_month_dong = df.groupby(['STD_YM', 'HDONG_NM', 'is_holiday']).mean().reset_index()
flow_time_df_month_dong
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(15, 5))

for i, row in flow_time_df_month_dong[flow_time_df_month_dong['is_holiday']==False].iterrows():
    #ts = row['TMST_00':'TMST_23'] / max(row['TMST_00':'TMST_23'])
    ts = np.log(list(row['TMST_00':'TMST_23']))# / max(row['TMST_00':'TMST_23'])
    ts = ts - min(ts)
    ts = ts / max(ts)
    plt.plot(ts, alpha=0.2)
    
plt.xticks(np.arange(24))
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15, 5))

for i, row in flow_time_df_month_dong[flow_time_df_month_dong['is_holiday']==False].iterrows():
    #ts = row['TMST_00':'TMST_23'] / max(row['TMST_00':'TMST_23'])
    ts = (row['TMST_00':'TMST_23'])# / max(row['TMST_00':'TMST_23'])
    ts = ts - min(ts)
    ts = ts / max(ts)
    plt.plot(ts, alpha=0.2)
    
plt.xticks(np.arange(24))
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15, 5))

for i, row in flow_time_df_month_dong[flow_time_df_month_dong['is_holiday']==True].iterrows():
    ts = np.log(list(row['TMST_00':'TMST_23']))
    plt.plot(ts, alpha=0.2)
    
plt.xticks(np.arange(24))
plt.show()
```

```python
long_flow_df = pd.wide_to_long(flow_time_df_month_dong, 'TMST_', i=['STD_YM', 'HDONG_NM', 'is_holiday'], j='time').reset_index()
long_flow_df['log(flow)'] = np.log(long_flow_df['TMST_'])
long_flow_df['year'] = long_flow_df['STD_YM'] // 100

long_flow_df
```

```python
import seaborn as sns

sns.set(style="darkgrid")
fig, ax = plt.subplots(figsize=(15, 5))

sns.lineplot(x='time', y='log(flow)', hue='is_holiday', style='year', data=long_flow_df)

plt.xticks(np.arange(24))
plt.show()
```

```python
lines = []

for i, row in flow_time_df_month_dong.iterrows():
    # 아침 4-8시
    t1_flow = np.log(row['TMST_04':'TMST_07'].mean())
    
    # 낮 8시-18시
    t2_flow = np.log(row['TMST_08':'TMST_17'].mean())
    
    # 저녁 18시-22시
    t3_flow = np.log(row['TMST_18':'TMST_21'].mean())
    
    # 밤 22시-4시
    t4_flow = np.log(np.mean(np.hstack((row['TMST_22':'TMST_23'], row['TMST_00':'TMST_03']))))
    
    lines.append((row['STD_YM'], row['HDONG_NM'], row['is_holiday'], t1_flow, t2_flow, t3_flow, t4_flow))
    
preproc_flow_df = pd.DataFrame(lines, columns=['STD_YM', 'HDONG_NM', 'is_holiday', 't1_flow', 't2_flow', 't3_flow', 't4_flow'])
preproc_flow_df.head()
```

```python
preproc_flow_df.to_pickle('2020빅콘테스트 문제데이터(혁신아이디어분야)/01_유동인구데이터(SK텔레콤)/preproc_flow_time_df.pkl')
```
